```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv('Heart Disease data.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.tail()
```

```python
summary_stats = df.describe()
print(summary_stats)

# Count of target variable (0: No disease, 1: Disease)
count_target = df['target'].value_counts()
print(count_target)

# Mean of continuous variables by target variable
mean_by_target = df.groupby('target').mean()
print(mean_by_target)
```

# Distribution of Age:

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='age', kde=True, color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

# Distribution of Cholesterol by target variable:
 

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='chol', hue='target', kde=True, color='skyblue')
plt.title('Distribution of Cholesterol by Target Variable')
plt.xlabel('Cholesterol')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Pairplot to show relationships between continuous variables

```python
sns.pairplot(data=df, hue='target', vars=['age', 'trestbps', 'chol', 'thalach'])
plt.title('Pairplot of Continuous Variables by Target Variable')
plt.show()
```

# Heatmap to show correlation between variables:

```python
plt.figure(figsize=(10, 8))
sns.heatmap(data=df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation between Variables')
plt.show()
```

# Boxplot to show distribution of continuous variables by target variable:


```python
sns.boxplot(x='target', y='trestbps', data=df)
plt.title('Distribution of Trestbps by Target Variable')
plt.xlabel('Target Variable')
plt.ylabel('Trestbps')
plt.show()
```

# Violinplot to show distribution of continuous variables by target variable:

```python
sns.violinplot(x='target', y='thalach', data=df)
plt.title('Distribution of Thalach by Target Variable')
plt.xlabel('Target Variable')
plt.ylabel('Thalach')
plt.show()
```

```python
Countplot to show distribution of categorical variables by target variable:
```

```python
sns.countplot(x='target', hue='sex', data=df)
plt.title('Distribution of Sex by Target Variable')
plt.xlabel('Target Variable')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
Barplot to show distribution of categorical variables by target variable:
```

```python
sns.barplot(x='target', y='cp', data=df, estimator=lambda x: x.mean())
plt.title('Mean of Cp by Target Variable')
plt.xlabel('Target Variable')
plt.ylabel('Mean')
plt.show()
```

```python
Scatterplot to show relationship between two continuous variables by target variable:
```

```python
sns.scatterplot(x='age', y='chol', hue='target', data=df)
plt.title('Relationship between Age and Chol by Target Variable')
plt.xlabel('Age')
plt.ylabel('Chol')
plt.legend()
plt.show()
```

```python
Swarmplot to show relationship between two categorical variables:
```

```python
sns.swarmplot(x='sex', y='exang', data=df)
plt.title('Relationship between Sex and Exang')
plt.xlabel('Sex')
plt.ylabel('Exang')
plt.show()
```

```python
Stripplot to show relationship between two categorical variables:
```

```python
sns.stripplot(x='cp', y='slope', data=df)
plt.title('Relationship between Cp and Slope')
plt.xlabel('Cp')
plt.ylabel('Slope')
plt.show()
```

```python
Boxenplot to show distribution of continuous variables by categorical variable:
```

```python
sns.boxenplot(x='thal', y='thalach', data=df)
plt.title('Distribution of Thalach by Thal')
plt.xlabel('Thal')
plt.ylabel('Thalach')
plt.show()
```

```python
Distplot to show distribution of continuous variables:
```

```python
sns.distplot(df['age'])
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
Boxplot to show distribution of continuous variables by categorical variable:
```

```python
sns.boxplot(x='fbs', y='chol', data=df)
plt.title('Distribution of Chol by Fbs')
plt.xlabel('Fbs')
plt.ylabel('Chol')
plt.show()
```
